```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def print_cumulative_mean_convergence(path_csv, print_csv=False):
    """
    Plotting the convergence of the cumulative mean for all columns in a CSV file.
    """
    df = pd.read_csv(path_csv)
    df = df.set_index('run number')

    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # calculate cumulative mean 
    df_cumulative_mean = df.expanding().mean()
    df_cumulative_mean = df_cumulative_mean.rename(columns=lambda x: f'{x}_Cumulative_Mean')
    
    if print_csv:
        print(df_cumulative_mean)
        
    # exctracting last row = final mean
    final_means = df_cumulative_mean.iloc[-1]
    print(final_means.round(2))

    for column in df_cumulative_mean.columns:
        plt.plot(df_cumulative_mean.index, df_cumulative_mean[column], 
                label=column)
        
    plt.title('Convergence of the Cumulative Mean for All Columns')
    plt.xlabel('Run Number')
    plt.ylabel('Cumulative Mean Value')
    plt.legend(title='Column', loc='best')
    plt.plot()
```

```python
print_cumulative_mean_convergence("./results/computation times/5x5-1_250k10k.csv")
```
